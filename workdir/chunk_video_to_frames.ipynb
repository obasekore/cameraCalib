```python
# grabing video and write it's frames as images to file
workdir = "C:/Users/OBAS/Pictures/Carrying_motion"
vid = cv2.VideoCapture("C:/Users/OBAS/Pictures/Carrying_motion/Media1.mp4")
cnt = 0
while(vid.isOpened()):
    ret, frame = vid.read()
    if ret == True:
        
        if((cnt%5)==0):
            cv2.imwrite(workdir+'/image_{0}_{0}.jpg'.format(cnt),frame)
            print("Processing Image ",cnt)
        cnt+=1
    else:
        break
```

```python
# sep = os.sep
working_dir = 'C:/Users/OBAS/Documents/Python_Scripts/Python_Scripts/RL_pybullet/gym-nao/gym_nao/envs/workdir/data_CoppeliaSimNAO'
video_extension = '.mp4'
cnt = 0
# list of all files & folder
for file in os.listdir(working_dir):
#     check for desired video extension
    if file.endswith(video_extension):
        
        path_file = os.path.join(working_dir,file)
        
        #create folder with video name
        folder_name = path_file.split('.')[0] 
        if (os.path.isdir(folder_name))==False:
            os.mkdir(folder_name)
        print(folder_name)
        print('****************************')
        # read the video
        vid = cv2.VideoCapture(working_dir+'/'+file)
        i = 0
        while(vid.isOpened()):
            ret, frame = vid.read()
            if ret == True:
                # print('Mean of frame %i is %1.1f' % (i, im.mean()))
                cv2.imwrite(os.path.join(folder_name,'{}.png'.format(cnt)),frame) # write each frame to file
                i +=1
                cnt +=1
            else:
                print('Error: No frame read')
                break
        print('{} images extracted'.format(i))
```
